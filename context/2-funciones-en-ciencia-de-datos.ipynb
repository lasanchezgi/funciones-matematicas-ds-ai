# Funciones en ciencia de datos

## Funciones de activación

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-10,10,N)
```

### Función lineal

$$y=mx+b$$


```python
def f(x):
    return x

plt.plot(x, f(x))
plt.grid()
plt.show()
```

### Función escalón o de Heaviside
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    
N=1000

y = H(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

### Función sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

```python
def f(x):
  return 1/(1 + np.exp(-x))

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

### Función tangente hiperbólica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

```python
def f(x):
  return np.tanh(x)
    
N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

### Función ReLU

$$R(x)=max(0,x)$$

```python
def f(x):
  return np.maximum(x,0)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
plt.show()
```
